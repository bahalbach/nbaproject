```python
seasons = [
    "2000-01",
    "2001-02",
    "2002-03",
    "2003-04",
    "2004-05",
    "2005-06",
    "2006-07",
    "2007-08",
    "2008-09",
    "2009-10",
    "2010-11",
    "2011-12",
    "2012-13",
    "2013-14",
    "2014-15",
    "2015-16",
    "2016-17",
    "2017-18",
    "2018-19",
    "2019-20",
    "2020-21"
]
for season_name in seasons[:1]:
    nbaTracker.load_season(season_name)
    teams = nbaTracker.current_season.team_seasons.values()
    season = nbaTracker.current_season
    missing = True
    while missing:
        missing = False
        for team in teams:
            if not hasattr(team, 'roster'):
                try:
                    roster = CommonTeamRoster(
                        team.id, season=season.name, timeout=60)
                    team.roster_df = roster.get_data_frames()[0]
                    team.coaches_df = roster.get_data_frames()[1]
                    print('adding', team.id)
                    time.sleep(5)
                except:
                    missing = True
    season.VERSION = 2
    season.save()
    print('saved season', season_name)
```
